###Desafio 003:
Proporção das classes tratamento?

```python
# Sem gráfico
dados['tratamento'].value_counts('com_droga', 'com_controle') * 100
```

```python
# Com gráfico
dados['tratamento'].value_counts('com_droga', 'com_controle') * 100 

# Criando conjunto de dados   
labels = 'com_droga', 'com_controle'
sizes = [dados['tratamento'].value_counts('com_droga', 'com_controle') * 100 ]

# Criando representação, área de plotagem
fig1, ax1 = plt.subplots()

# Criando o gráfico
ax1.pie(sizes, labels = labels, autopct = '%1.1f%%',
        shadow = True, startangle = 90)

# Gŕafico em círculo
ax1.axis('equal')

# Mostra o gráfico
plt.show()
```

###Desafio 007:
Deixar o gráficos bonitões. (Matplotlib.pyplot)


```python
#dados['droga'].value_counts().head().plot.pie()

#Criando conjunto de dados
labels = dados['droga'].value_counts().head()
sizes = [dados['droga'].value_counts().head()]

# Criando representação, área de plotagem

fig1, ax1 = plt.subplots()

# Criando o gráfico
plt.title('5 drogas mais utilizadas')
ax1.pie(sizes, labels = labels, autopct = '%1.1f%%',
        shadow = True, startangle = 90)

# Gŕafico em círculo
ax1.axis('equal')

# Mostra o gráfico
plt.show()

```
