```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

url_dados = 'https://github.com/alura-cursos/imersaodados3/blob/main/dados/dados_experimentos.zip?raw=true'

dados = pd.read_csv(url_dados, compression = 'zip')
dados
```

#Desafio 01: Ordenar o gráfico com countplot.

```python
#Renomear a coluna 'droga para 'composto'
mapa = {'droga': 'composto'}
dados.rename(columns = mapa, inplace = True)

#Cinco drogas utilizadas com mais frequencia
cod_composto = dados['composto'].value_counts().index[0:5]
```

```python
#Gráfico de barras com informações dos 5 compostos com maior frequencia de utilização, com a biblioteca seaborn
sns.set()
plt.figure(figsize = (8, 6))
ax = sns.countplot(x = 'composto', data = dados.query('composto in @cod_composto'))
ax.set_title('Top 5 Compostos')
plt.show()
```

```python
#Gráfico de barras com informações dos 5 compostos com maior frequencia de utilização, com a biblioteca seaborn, ordenado da maior coluna para a menor
sns.set()
plt.figure(figsize = (8, 6))
ax = sns.countplot(x = 'composto', order = ['cacb2b860', '87d714366', '9f80f3f77', '9f80f3f77', '8b87a7a83'] , data = dados.query('composto in @cod_composto'))
ax.set_title('Top 5 Compostos')
plt.show
```

#Desafio 02: Melhorar a visualização do gráfico, alterando o tamanho da fonte, mudando o eixo y de 'count' para 'contagem', deixando o título mais visível, etc.

```python
#Gráfico de barras com informações dos 5 compostos com maior frequencia de utilização, com a biblioteca seaborn, 
#ordenado da maior coluna para a menor, com outro tamanho de fonte, mudando o nome do eixo y e deixando o título mais vísivel.
sns.set()
plt.figure(figsize = (8, 6))
plt.rcParams['xtick.labelsize'] = 12
plt.rcParams['ytick.labelsize'] = 12
ax = sns.countplot(x = 'composto', order = ['cacb2b860', '87d714366', '9f80f3f77', '9f80f3f77', '8b87a7a83'] , data = dados.query('composto in @cod_composto'))
ax.set_ylabel('contagem')
ax.set_title('Top 5 Compostos', fontsize = 18)
plt.show
```

#Desafio 04: Estudar sobre as estatísticas retornadas no método .describe().

count: Retorna a quantidade total de valores não nulos da base de dados selecionada.
<br>
mean: Retorna a média dos valores somados da base de dados selecionado.
<br>
std: Retorna o cálculo do desvio padrão da soma dos valores da base de dados selecionada.
<br>
25%: Retorna a média do 1º quarto de valores(os primeiros 25%) da base de dados selecionada.
<br>
50%: Retorna média dos valores da mediana(os valores que se encontram no meio da base de dados) da base de dados selecionada.
<br>
75%: Retorna a média do 3º quarto de valores(os 25% após a mediana) da base de dados selecionada.
<br>
max: Retorna o maior valor da base de dados selecionado
